# Floor-Plan Loader Check

This notebook loads a traced PNG floor plan into the locked tri-state grid model and displays it with Matplotlib.

```python
from pathlib import Path
from src.common.floorplan_loader import load_traced_floorplan
```

```python
TRACED_floorplan_PATH = Path("../../static/floor-plan/traced/ground-back.png")

if not TRACED_floorplan_PATH.exists():
    raise FileNotFoundError(
        f"Expected traced floor-plan PNG at {TRACED_floorplan_PATH}. Start the notebook from somewhere inside the repository."
    )

TRACED_floorplan_PATH
```

```python
floorplan = load_traced_floorplan(TRACED_floorplan_PATH)
floorplan
```

```python
axis = floorplan.plot(title=f"Loaded Grid: {floorplan.name}")
axis.figure.set_size_inches(12, 5) # type: ignore
```

```python
print(f"shape: {floorplan.shape}")
print(f"null cells: {floorplan.null_cell_count}")
print(f"open cells: {floorplan.open_cell_count}")
print(f"solid cells: {floorplan.solid_cell_count}")
print(f"grid_cell_size_m: {floorplan.grid_cell_size_m}")
```
